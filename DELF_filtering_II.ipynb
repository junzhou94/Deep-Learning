```python
import tensorflow as tf
import tensorflow_hub as hub

# Prepare an image tensor.
def image_input_fn(path):
    image_contents = tf.read_file(path)
    image_tf = tf.image.decode_jpeg(image_contents, channels = 3)
    return tf.image.convert_image_dtype(image_tf, tf.float32)

# Instantiate the DELF module.
delf_module = hub.Module("https://tfhub.dev/google/delf/1")
```

```python
image_placeholder = tf.placeholder(tf.float32, shape=(None, None, 3), name='input_image')
delf_inputs = {
  # An image tensor with dtype float32 and shape [height, width, 3], where height and width are positive integers:
  'image': image_placeholder,
  # Scaling factors for building the image pyramid as described in the paper:
  'image_scales': [0.25, 0.3536, 0.5, 0.7071, 1.0, 1.4142, 2.0],
  # Image features whose attention score exceeds this threshold will be returned:
  'score_threshold': 100.0,
  # The maximum number of features that should be returned:
  'max_feature_num': 500,
}

# Apply the DELF module to the inputs to get the outputs.
delf_outputs = delf_module(delf_inputs, as_dict = True)
```

```python
from skimage.transform import AffineTransform
from skimage.measure import ransac
from scipy.spatial import cKDTree

test_revised = pd.read_csv('test_revised_based_on_20.csv')
#train_sample_2 = pd.read_csv('train_sample_2.csv')
#train_sample_3 = pd.read_csv('train_sample_3.csv')
train_sample_4 = pd.read_csv('train_sample_4.csv')
#train_sample_5 = pd.read_csv('train_sample_5.csv')
test_revised.shape
```

```python
test_revised = test_revised.sort_values(by = 'confidence', ascending = False)
test_revised = test_revised.sort_values(by = 'landmark_id', ascending = True)
test_revised = test_revised.reset_index().drop('index', axis = 1)
```

```python
iters = 14951   # 14951, means landmark_id
n = 4

rec = list()
    
for i in range(iters):
    
    print(i)
    
    img_id = test_revised[test_revised['landmark_id'] == i]['id']
    if(img_id.shape[0] == 0):
        continue
    
    cmp_list = list()
    cmp_img_id = train_sample_4[train_sample_4['landmark_id'] == i]['id']
    for j in range(min(n, cmp_img_id.shape[0])):

        current_path = '../data/train/' + cmp_img_id.iloc[j] + '.jpg'
        
        if os.path.exists(current_path): 
            
            image_tf = image_input_fn(current_path)        
        
            tmp_dict = dict()
            with tf.train.MonitoredSession() as sess:        
                image = sess.run(image_tf)
                tmp_dict[cmp_img_id.iloc[j]] = sess.run([delf_outputs['locations'], delf_outputs['descriptors']], feed_dict = {image_placeholder: image})
        
            cmp_list.append(tmp_dict)
        
        else:
            cmp_list.append(-1)
    
    res_list = list()
    for j in range(img_id.shape[0]):
        current_path = '../data/test/' + img_id.iloc[j] + '.jpg'
        image_tf = image_input_fn(current_path) 
    
        tmp_dict = dict()
        with tf.train.MonitoredSession() as sess:        
            image = sess.run(image_tf)
            tmp_dict[img_id.iloc[j]] = sess.run([delf_outputs['locations'], delf_outputs['descriptors']], feed_dict = {image_placeholder: image})
        
        res_list.append(tmp_dict)
        
        # calculation
        
        inliers = [0]
        #avg_inliers = [0]
        inliers_2 = [0]
        #avg_inliers_2 = [0]
        
        locations_1, descriptors_1 = res_list[j][img_id.iloc[j]]
        num_features_1 = locations_1.shape[0]
        
        distance_threshold = 0.9
        d1_tree = cKDTree(descriptors_1)
        
        for k in range(len(cmp_list)):
            
            if cmp_list[k] != -1:

                locations_2, descriptors_2 = cmp_list[k][cmp_img_id.iloc[k]]
                num_features_2 = locations_2.shape[0]
            
                _, indices = d1_tree.query(descriptors_2, distance_upper_bound = distance_threshold)

                locations_to_use = np.array([
                    locations_2[i,]
                    for i in range(num_features_2)
                    if indices[i] != num_features_1
                ])
        
                inliers.append(locations_to_use.shape[0])
                inliers_2.append(pd.unique(indices).shape[0])
        
        if len(inliers) > 1:
            avg_inliers = sum(inliers) / (len(inliers) - 1)
        else:
            avg_inliers = 0
        if len(inliers_2) > 1:
            avg_inliers_2 = sum(inliers_2) / (len(inliers_2) - 1)
        else:
            avg_inliers_2 = 0
        
        rec.append([max(inliers), avg_inliers, max(inliers_2), avg_inliers_2])
```

```python
pd.DataFrame(rec).to_csv('labels_DELF2_0_14951.csv', index = False)
```
